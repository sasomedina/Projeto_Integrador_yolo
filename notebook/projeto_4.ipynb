```python
gsc = GridSearchCV(
estimator=RandomForestRegressor(random_state=42),
param_grid={'max_depth': range(3,7),
            'n_estimators': (10, 50, 100, 200, 300, 400)},
        cv=5, scoring='r2', verbose=0, n_jobs=-1)
    
grid_result = gsc.fit(X_train, y_train)
best_params = grid_result.best_params_
print(best_params)
    
rfr = RandomForestRegressor(max_depth=best_params["max_depth"], n_estimators=best_params["n_estimators"],random_state=42, verbose=False)

# Validação cruzada
scores = cross_val_score(rfr, X_train, y_train, cv=10, scoring='r2')
print(scores)
```
